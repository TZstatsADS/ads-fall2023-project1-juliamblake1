## Work vs. Love for men vs. women

Now we will use a method called word2vec to find words most associated with 'work' for men and women. First, we must train the models individually for the full, men and women datasets.

```python
import itertools
res = list(itertools.filterfalse(lambda item: not item , txt)) # remove None produced

sent = []
for piece in res:
    sent.append(piece.split())

phrases = Phrases(sent, min_count=30, progress_per=10000)
bigram = Phraser(phrases)
sentences = bigram[sent]
trigram = Phrases(sentences)
```

```python
cores = multiprocessing.cpu_count() # Count the number of cores in a computer
w2v_model = Word2Vec(min_count=20, # load a pretrained word2vec model
                     window=2,
                     vector_size=300,
                     sample=6e-5, 
                     alpha=0.03, 
                     min_alpha=0.0007, 
                     negative=20,
                     workers=cores-1) 
```

```python
w2v_model.build_vocab(sentences, progress_per=10000)
w2v_model.train(sentences, total_examples=w2v_model.corpus_count, epochs=30, report_delay=1)
```

```python
w2v_model.wv.most_similar(positive=["work"])
```

Now we will look at 'work' for men.

```python
res = list(itertools.filterfalse(lambda item: not item , male_txt)) # remove None produced

sent = []
for piece in res:
    sent.append(piece.split())
    
male_sentences = bigram[sent]

w2v_model_men = Word2Vec(min_count=20, # load a pretrained word2vec model
                     window=2,
                     vector_size=300,
                     sample=6e-5, 
                     alpha=0.03, 
                     min_alpha=0.0007, 
                     negative=20,
                     workers=cores-1) 

w2v_model_men.build_vocab(male_sentences, progress_per=10000)
w2v_model_men.train(male_sentences, total_examples=w2v_model.corpus_count, epochs=30, report_delay=1)


w2v_model_men.wv.most_similar(positive=["work"])
```

```python
res = list(itertools.filterfalse(lambda item: not item , female_txt)) # remove None produced

sent = []
for piece in res:
    sent.append(piece.split())
    
female_sentences = bigram[sent]

w2v_model_female = Word2Vec(min_count=20, # load a pretrained word2vec model
                     window=2,
                     vector_size=300,
                     sample=6e-5, 
                     alpha=0.03, 
                     min_alpha=0.0007, 
                     negative=20,
                     workers=cores-1) 

w2v_model_female.build_vocab(female_sentences, progress_per=10000)
w2v_model_female.train(female_sentences, total_examples=w2v_model.corpus_count, epochs=30, report_delay=1)


w2v_model_female.wv.most_similar(positive=["work"])
```

When we look at the words most related to 'work' within the subsets of men and women we also see a different trend emerge. Men and women both refer to their boss, manager or supervisor, however, second most related to work for women was 'quit' and fifth was'receive_bonus'. Whereas men were more focused on potentially taking shifts and assigning work and completing overdue tasks.

## Conclusion

In conclusion, the contrasting emphasis on work over affection in the descriptions provided by men reveals a significant aspect of their priorities and how they choose to articulate their sources of happiness. The words they employ to depict their experiences underscore a strong orientation towards professional pursuits and accomplishments.

Throughout the narratives, men frequently emphasized words like "work," "career," "achievement," and other related terms. These expressions reflect a deep-seated dedication to their professional roles and a sense of fulfillment derived from making progress in their careers. The choice of words emphasizes their commitment to the workplace, positioning their achievements as central sources of happiness.

In contrast, expressions of affection, while present, often played a secondary role in the narratives of men. Affection-related words such as "love," "romance," and "emotion" were not as prominent as terms associated with work. This pattern of expression suggests that, for many men, professional success and the pursuit of career goals occupy a more prominent place in their lives.

It is important to recognize that these findings are based on observed trends within the dataset and may not encompass the full spectrum of individual experiences. Furthermore, societal expectations and gender roles are evolving, and many individuals prioritize both work and affection in their lives in diverse ways.

In summary, the inclination of men to focus more on work-related terms in their descriptions of happy moments underscores the significance of their professional identities and achievements. This preference for work-related language reflects the multifaceted nature of happiness and the individualized ways in which people articulate and prioritize the sources of their joy.
